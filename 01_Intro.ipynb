## Hvad er programmering? En rejse ind i den digitale verden!

Hej fremtidige udviklere! Har i nogensinde tænkt over, hvordan jeres yndlingsvideospil bliver lavet, hvordan apps på din telefon ved, hvad de skal gøre, eller hvordan hjemmesider som YouTube eller TikTok fungerer? Hemmeligheden bag alle disse fantastiske digitale kreationer er noget, der kaldes **programmering**!

### Hvad er programmering?

Forestil dig programmering som at give en computer instrukser. Computere er super smarte, men de kan ikke tænke selv. De har brug for, at vi fortæller dem præcis, hvad de skal gøre, skridt for skridt. Når vi programmerer, skriver vi disse instrukser i specielle sprog, som computere kan forstå, som f.eks. Python, Java eller Scratch. Det er ligesom at skrive en opskrift, som en computer skal følge!

### En lille smule historie

Programmering er ikke en ny idé! Folk har forsøgt at automatisere opgaver i århundreder. Men den virkelige magi startede med pionerer som **Ada Lovelace** i 1800-tallet, som ofte kaldes den første computerprogrammør. Hun forstod, at maskiner kunne gøre mere end bare at regne tal – de kunne følge komplekse instrukser. Spol frem til 1940'erne og 50'erne, og vi fik de allerførste elektroniske computere. Disse tidlige computere var enorme og krævede, at programmører skulle flippe kontakter og sætte ledninger i for at give dem instrukser! Med tiden blev programmeringssprogene mere lig menneskelige sprog, hvilket gjorde det lettere for os at kommunikere med computere.

### Hvor ser vi programmering?

Programmering er overalt omkring os! Det er ikke kun for computervidenskabsfolk i laboratorier. Her er et par eksempler:

*   **Videospil:** Hver eneste karakterbevægelse, hver speciel effekt, hver pointberegning i spil som Minecraft eller Fortnite er takket være programmering.
*   **Smartphone-apps:** Fra Snapchat-filtre til Google Maps, Instagram og endda lommeregneren på din telefon – apps er bygget med kode.
*   **Hjemmesider:** Når du surfer på internettet, ser videoer på YouTube eller søger på Google, arbejder programmering i kulisserne for at vise dig alt.
*   **Robotter og smarte enheder:** Din smarthøjttaler, en robotstøvsuger eller endda trafiklysene ved et vejkryds styres af programmer.
*   **Film og animation:** Specialeffekter i superheltefilm er ofte skabt ved hjælp af kompleks programmering.

At lære at programmere er som at få en superkraft – du får mulighed for at skabe, bygge og løse problemer i den digitale verden! Det er en færdighed, der kan åbne op for uendelige muligheder for din fremtid.

### Velkommen til Google Colaboratory (Colab)!

Google Colaboratory, eller bare Colab, er et fantastisk værktøj, der lader os skrive og køre Python-kode direkte i din webbrowser. Det er som en digital notesbog, hvor vi både kan skrive forklaringer (som denne tekst) og køre vores kode. Det er helt gratis og perfekt til at lære at programmere!

#### Hvordan bruger man Colab?

*   **Opret en ny notesbog**: Du kan altid starte en frisk notesbog ved at gå op til menuen og vælge `Filer > Ny notesbog` (File > New notebook). Så får du et blankt lærred til din kode.

*   **Naviger i brugerfladen**: En Colab notesbog er bygget op af små kasser, der kaldes 'celler'. Der er to hovedtyper:
    *   **Tekstceller (som denne!)**: Disse bruges til forklaringer, instruktioner og noter. Du kan redigere dem ved at dobbeltklikke.
    *   **Kodeceller**: Disse indeholder den faktiske Python-kode, vi vil køre. Du vil se en afspilningsknap (▶) ved siden af dem.

*   **Kør kodeceller**: For at få computeren til at udføre koden i en celle, kan du gøre en af disse ting:
    *   Klik på **afspilningsknappen** (▶), der dukker op, når du holder musen over cellen.
    *   Tryk på `Shift + Enter` på dit tastatur, mens cellen er valgt.

Lad os prøve at køre en simpel kodecelle lige nedenfor!

```python
print('Hej, Colab!')
```

## Vores Programmeringssprog: Python!

Nu hvor du ved, hvad programmering er, og hvordan du bruger Colab, lad os dykke ned i den spændende verden af **Python**! Python er et af de mest populære programmeringssprog i verden, og det er det, vi vil bruge til at lære at kode.

### Hvad er Python?

Python er et alsidigt programmeringssprog, hvilket betyder, at det kan bruges til mange forskellige ting. Det blev skabt af en programmør ved navn Guido van Rossum i slutningen af 1980'erne. Et af dets hovedmål var at være let at læse og skrive, selv for begyndere!

### Hvorfor Python er fantastisk for begyndere

1.  **Læsevenlighed**: Python-kode ligner ofte almindelig engelsk. Dette gør det lettere at forstå, hvad der sker i dit program, især når du lige er startet.
2.  **Enkelhed**: Python har en ren og ligetil syntaks, hvilket betyder, at reglerne for at skrive kode er nemme at lære. Du behøver ikke at bekymre dig om en masse komplicerede symboler eller forvirrende strukturer.
3.  **Stort Fællesskab**: Mange mennesker bruger Python, så der er masser af ressourcer, tutorials og fællesskaber online, der kan hjælpe dig, hvis du går i stå eller vil lære mere.

### Almindelige anvendelsesmuligheder for Python

Python er overalt! Her er blot nogle få eksempler på, hvor Python bruges:

*   **Webudvikling**: Bygning af hjemmesider og webapplikationer (som Instagram eller YouTube).
*   **Dataanalyse og Videnskab**: Hjælper forskere og virksomheder med at forstå enorme mængder data.
*   **Kunstig Intelligens (AI) & Maskinlæring**: Driver intelligente systemer som anbefalingsmotorer (tænk Netflix-forslag!) eller selvkørende biler.
*   **Spiludvikling**: Skabelse af videospil.
*   **Automatisering**: Får computere til at udføre gentagne opgaver for os automatisk.

### Grundlæggende Programmeringssyntaks: Præcise instruktioner

Husker du, hvordan vi sagde, at programmering handler om at give instruktioner til en computer? Ligesom når du skriver en opskrift eller giver vejbeskrivelser, skal du være meget præcis. Computere forstår ikke slang eller vage kommandoer. Hvert programmeringssprog har sin egen **syntaks**, som er ligesom grammatikken og ordforrådet i det sprog.

For eksempel ved vi på dansk, at en sætning normalt starter med stort bogstav og slutter med et punktum. I Python er der også regler for, hvordan du skriver kommandoer, bruger tegnsætning og arrangerer din kode. Hvis du får syntaksen forkert, vil computeren ikke forstå dine instruktioner, og dit program fungerer måske ikke som forventet. Vi vil lære Pythons syntaks trin for trin, så vores computervenner forstår os perfekt!

### Fra "klodser" til "kode": Python og Scratch/Blockly

Hvis du har arbejdet med Scratch eller Blockly før, så har du allerede en rigtig god forståelse for, hvad programmering handler om! I disse programmeringsmiljøer byggede du dine programmer ved at trække og slippe farverige "klodser" sammen. Hver klods repræsenterede en instruktion, og når du satte dem sammen, fortalte du computeren, hvad den skulle gøre.

Python er lidt anderledes – her skriver vi instruktionerne som **tekst** i stedet for at bruge klodser. Men principperne er præcis de samme! Tænk på det på denne måde:

*   **Klodser i Scratch/Blockly** = **Linjer af kode i Python**
*   **At sætte klodser sammen i den rigtige rækkefølge** = **At skrive Python-instruktioner i den rigtige rækkefølge**

For eksempel, hvis du i Scratch havde en klods, der sagde "sige Hej!", så ville det i Python typisk se ud som `print('Hej!')`. Begge dele gør det samme: de får computeren til at vise teksten "Hej!".

Fordelen ved tekstbaseret programmering som Python er, at det giver dig meget mere fleksibilitet og magt til at skabe mere komplekse programmer. Men det grundlæggende "tænk-som-en-programmør" er det samme, uanset om du bruger klodser eller tekst!

## Output med `print()`

### Hvad er `print()`?

I programmering er det ofte vigtigt at kunne se, hvad vores program laver, eller at vise information til brugeren. Det er her `print()`-funktionen kommer ind i billedet!

`print()` er en indbygget funktion i Python, der bruges til at vise tekst, tal eller resultatet af beregninger på skærmen (som regel i den udgang, du ser nedenfor).

### Hvordan bruger man `print()`?

Det er supernemt! Du skriver bare `print` efterfulgt af det, du vil vise, inde i parenteser `()`.

*   **Tekst (strenger)**: Hvis du vil vise tekst, skal du omslutte teksten med enkelte ('') eller dobbelte (" ") citationstegn. Python kalder dette for en 'streng' (string).
    Eksempel: `print('Hej verden!')`

*   **Tal**: Hvis du vil vise tal, skriver du bare tallet direkte inde i parenteserne.
    Eksempel: `print(10)`

*   **Regnestykker**: Du kan også bede `print()` om at vise resultatet af et regnestykke. Python vil først regne ud, hvad regnestykket giver, og derefter vise svaret.
    Eksempel: `print(5 + 3)`

Lad os prøve nogle eksempler!

```python
print('Dette er en tekststreng!')
```

```python
print(123)
```

```python
print(5 + 3 * 2)
```

### Hvad er det næste? Introduktion til Variable!

Forestil dig, at du vil gemme en information, f.eks. dit navn eller et tal, som du vil bruge senere i dit program. Hvor lægger du det? I programmering bruger vi noget, der hedder **variable** til dette!

Variable er som små navngivne kasser eller beholdere, hvor du kan opbevare forskellige typer information – tekst, tal, og meget mere. Du kan lægge noget i kassen, tage det ud, eller endda ændre det indhold, der er i kassen. Dette gør dine programmer meget mere fleksible og kraftfulde!

I næste lektion skal vi dykke ned i variable og lære, hvordan vi kan bruge dem til at gemme og manipulere data i vores Python-programmer. Det bliver en vigtig byggeklods for alle de seje ting, du vil kunne skabe!
