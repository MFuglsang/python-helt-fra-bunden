# Løkker - Gentag Opgaver Automatisk

I denne arbejdsbog skal du lære at bruge løkker. En løkke er en måde at få computeren til at gentage en opgave mange gange uden at skulle skrive den samme kode igen og igen.

Forestil dig at du skal tegne 100 stjerner. Du kunne tegne hver enkelt - eller du kunne sige til computeren: 'Tegn en stjerne 100 gange!' Det er præcis hvad en løkke gør.

Lad os tage det helt fra start! 🚀

## Hvad er en Løkke?

En løkke er som en opskrift med ordene "gentag dette".

### Eksempel fra virkeligheden:

**Uden løkke:**
```
Tegn en stjerne
Tegn en stjerne
Tegn en stjerne
Tegn en stjerne
Tegn en stjerne
...(100 gange!)
```

**Med løkke:**
```
Gentag 100 gange:
    Tegn en stjerne
```

Se hvor meget kortere det er! Det er det som programmer ofte skal gøre - gentage samme ting mange gange meget hurtigt.

### Hvornår bruger vi løkker?

Løkker er overalt i programmer! Her er nogle eksempler:

- **Spil**: Når et spil skal tegne 1000 fjender på skærmen bruger det en løkke
- **Sociale medier**: Når Facebook loader dine 500 venners profilbilleder bruger det en løkke
- **Musik-streaming**: Når Spotify skal afspille 50 mio. sange bruger den løkker til at behandle hver sang
- **Vejrudsendelser**: Når meteorologer skal analysere temperaturdata fra 100.000 vejrstationer bruger de løkker
- **Videospil-grafik**: Hvis skærmen har 2560x1440 pixels må computeren tegne alle 3.686.400 pixels - med en løkke!

**Uden løkker ville programmering være næsten umulig!** Enhver gang du skal gøre noget gentagne gange, er en løkke der til at hjælpe.

## Den Mest Brugte Løkke: `for` Løkken

Den mest almindelige løkke hedder `for`. Den bruges når du ved præcis hvor mange gange du skal gentage noget.

### Syntaks:

```python
for tal in range(5):
    print(tal)
```

Dette betyder: "For hver tal fra 0 til 4, udskriv tallet."

### Vigtige dele:
- `for` - starter en for-løkke
- `tal` - et navn vi vælger (kunne være `i`, `tal`, `nummer`, eller hvad du vil)
- `in range(5)` - betyder "fra 0 til 4" (5 tal i alt)
- `:` - kolon efter løkken (vigtig!)
- Koden under er *indrykket* (rykket ind) - det betyder det er "inde i" løkken

## Hvad er `range()`?

`range()` er en funktion som laver en sekvens af tal.

### Eksempler:

- `range(5)` → 0, 1, 2, 3, 4 (fem tal, starter fra 0)
- `range(1, 6)` → 1, 2, 3, 4, 5 (fra 1 til 5)
- `range(0, 10, 2)` → 0, 2, 4, 6, 8 (hver anden tal)

**Husk:** `range()` starter altid fra 0 (eller hvad du siger) og stopper ÉN før det sidste tal!

### Lad os Prøve det!

````python
# Eksempel 1: Udskriv tallene 0 til 4
for tal in range(5):
    print(tal)
````

**Hvad sker der?**

Computeren kører løkken 5 gange. Første gang er `tal = 0`, anden gang `tal = 1`, osv.

Hver gang udskriver vi tallet.

## Løkker med Beregninger - Gangetabel

Du kan bruge tallene fra løkken til at lave beregninger!

````python
# Eksempel 2: Gangetabellen for 5
print('Gangetabellen for 5:')
print('---')
for i in range(1, 11):  # Fra 1 til 10
    resultat = i * 5
    print(str(i) + ' * 5 = ' + str(resultat))
````

**Hvad sker her?**
- Vi løber fra 1 til 10
- Hver gang ganger vi tallet med 5
- Vi bruger `str()` til at konvertere tallene til tekst, så vi kan kombinere dem med `+`

**Vigtig:** Husk at `str()` konverterer tal til tekst så du kan kombinere dem!

**Prøv selv:** Hvad hvis vi ændrer `range(1, 11)` til `range(1, 13)`? Hvor mange linjer får vi?

## Løkker og Betingelser - Hvis Du Skal Tage Beslutninger

Du kan bruge `if` statements *inde i* en løkke! Husk hvad du lærte i Modul 03?

````python
# Eksempel 3: Find alle lige tal fra 1-10
print('Lige tal fra 1-10:')
for tal in range(1, 11):
    if tal % 2 == 0:  # % betyder "rest efter division" - hvis rest er 0, er det lige
        print(tal)
````

**Hvad sker her?**
- Løkken gør gennem tallene 1 til 10
- `tal % 2` giver resten hvis vi deler med 2
- Hvis resten er 0 (dvs. `tal % 2 == 0`), er tallet lige
- Vi udskriver kun de lige tal

**Vigtig:** Husk begge indentationer! `if` skal være rykket ind under `for`, og `print` skal være rykket ind under `if`!

## Løkker gennem Tekst (Strenge)

Du kan også løbe gennem hver bogstav i en streng!

````python
# Eksempel 4: Løbe gennem hver bogstav
navn = 'Python'
print('Bogstaver i navn:')
for bogstav in navn:
    print(bogstav)
````

**Hvad sker her?**
- `for bogstav in navn` betyder: "For hver bogstav i strengen 'Python'"
- Vi løber fra 'P' til 'n'
- Udskriver hver enkelt bogstav

**Husk:** Du kan løbe gennem en streng på samme måde som du løber gennem tal med `range()`!

## Tegn Mønster Med Løkker

Sjov kendsgerning: Du kan gange en streng med et tal!

````python
# Eksempel 5: Gentag en streng
print('*' * 5)  # Dette printer 5 stjerner
print('---' * 3)  # Dette printer 9 bindestreger (3 gange 3 tegn)
print('Ho' * 4)  # Dette printer 'HoHoHoHo'
````

**Hvad sker her?**
- `'*' * 5` betyder: "Gentag stjernen 5 gange"
- Det virker fordi Python kender forskel på at bruge `*` med tal vs. strenge
- Med tal: `5 * 3 = 15` (matematik)
- Med streng: `'*' * 5 = '*****'` (gentag)

## Fælles Fejl - Pas På!

### Fejl 1: Glemmer Indentation (Røkket ind)
```python
for i in range(5):
print(i)  # FEJL - ikke rykket ind!
```
**Løsning:** Ryk løkke-koden ind med 4 mellemrum eller en Tab

### Fejl 2: Glemmer Kolon
```python
for i in range(5)  # FEJL - mangler kolon!
    print(i)
```
**Løsning:** Læg altid `:` efter for-linjen

### Fejl 3: range() Starter fra 0, Ikke 1
```python
for i in range(5):
    print(i)  # Udskriver: 0, 1, 2, 3, 4 (ikke 1, 2, 3, 4, 5)
```
**Løsning:** Hvis du vil fra 1, skriv `range(1, 6)`

### Fejl 4: Glemmer str() Når Du Kombinerer Tal og Tekst
```python
tal = 5
print('Tallet er ' + tal)  # FEJL - kan ikke kombinere string og int!
```
**Løsning:** Brug `str(tal)` først
```python
print('Tallet er ' + str(tal))  # Virker!
```
