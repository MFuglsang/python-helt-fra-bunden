## Hvad er en Liste?

En liste er som en indkøbsliste - en samling af ting på række.

### Eksempel fra virkeligheden:

**Uden liste (gamle måde):**
```python
ven1 = 'Maria'
ven2 = 'Anders'
ven3 = 'Sofia'
ven4 = 'Lars'
```

**Med liste (bedre måde):**
```python
venner = ['Maria', 'Anders', 'Sofia', 'Lars']
```

Meget kortere og nemmere at arbejde med!

## Opret en Liste

Her er syntaksen:

```python
liste = [element1, element2, element3]
```

### Vigtige dele:
- `[` og `]` - kvadratparenteser markerer en liste
- Elementer adskilles med komma
- Du kan have tal, tekst, eller blanding!
- Tom liste: `[]`

### Lad os Prøve det!

````python
# Eksempel 1: Opret forskellige lister
venner = ['Maria', 'Anders', 'Sofia']
tal = [1, 2, 3, 4, 5]
blanding = ['Python', 10, 'programmer', 25]

print(venner)
print(tal)
print(blanding)
````

**Hvad sker der?**

Vi oprettede 3 lister:
- `venner` - liste af navne (strenge)
- `tal` - liste af tal (integers)
- `blanding` - blanding af strenge og tal

Når vi printer dem, viser Python hele listen med hakeparenteser.

## Adgang til Elementer - Indeksering

Du kan få adgang til et specifikt element i listen ved at bruge dets position (indeks).

**VIGTIG REGEL:** Python starter tælling fra 0, ikke 1!

```
liste = ['Maria', 'Anders', 'Sofia']
indeks:    0        1        2
```

````python
# Eksempel 2: Adgang til elementer
frugt = ['æble', 'banan', 'kirsebær', 'drue']

print(frugt[0])  # Første element
print(frugt[1])  # Anden element
print(frugt[2])  # Tredje element
print(frugt[3])  # Fjerde element
````

**Hvad sker der?**

- `frugt[0]` giver 'æble' (første element)
- `frugt[1]` giver 'banan' (anden element)
- osv.

**Vigtig:** Hvis du skriver `frugt[4]` får du fejl fordi der ikke er element nummer 4!

# Eksempel 3: Negative indeks
frugt = ['æble', 'banan', 'kirsebær', 'drue']

print(frukt[-1])  # Sidste element
print(frukt[-2])  # Næsten sidste element
print(frukt[-3])  # Tredje element bagfra

**Hvad sker her?**

- `frukt[-1]` giver sidste element
- `frukt[-2]` giver næsten sidste
- Det er praktisk når du vil have sidste element uden at vide listen' længde!

## Løkker gennem Lister

Du kan løbe gennem hver element i en liste - helt som du gjorde med strenge i Modul 04!

````python
# Eksempel 4: Løbe gennem liste
dyr = ['kat', 'hund', 'fugl', 'fisk']

print('Mine favoritdyr:')
for dyr_navn in dyr:
    print('- ' + dyr_navn)
````

**Hvad sker her?**

- `for dyr_navn in dyr:` betyder "For hver element i listen"
- Første iteration: `dyr_navn = 'kat'`
- Anden iteration: `dyr_navn = 'hund'`
- osv.

**Dette er meget kraftfuldt!** Du kan gøre noget med hvert element!

## Ændring af Lister - .append()

Du kan tilføje nye elementer til en liste med `.append()`!

````python
# Eksempel 5: Tilføj elementer
indkøbsliste = ['brød', 'meløj', 'æg']

print('Oprindelig liste:')
print(indkøbsliste)

# Tilføj nye elementer
indkøbsliste.append('smør')
indkøbsliste.append('ost')

print('\nEfter at have tilføjet:')
print(indkøbsliste)
````

**Hvad sker her?**

- `indkøbsliste.append('smør')` tilføjer 'smør' til SLUTNINGEN af listen
- Listen ændres - der er ingen `.append()` uden effekt
- Hver `.append()` tilføjer ét element

## Fjernelse af Elementer - .remove()

Du kan fjerne elementer med `.remove()`!

````python
# Eksempel 6: Fjern elementer
hvad_at_gøre = ['matematik', 'engelsk', 'idrætslektie', 'historie']

print('Oprindelig liste:')
print(hvad_at_gøre)

# Fjern element
hvad_at_gøre.remove('idrætslektie')

print('\nEfter at have fjernet "idrætslektie":')
print(hvad_at_gøre)
````

**Hvad sker her?**

- `hvad_at_gøre.remove('idrætslektie')` fjerner det element
- Listen ændres
- Der kan kun fjernes én ting ad gangen

## Længde af Liste - len()

Du kan finde ud af hvor mange elementer der er i en liste med `len()`!

**Hvad sker her?**

- `len(flader)` giver antal elementer (3)
- Vi konverterer det til streng med `str()`
- Vi kan også bruge `len()` i en løkke: `range(len(liste))` betyder "fra 0 til liste-længde"

## Slicing - Få del af Liste

Du kan få en del af listen med `liste[start:slut]`!

````python
# Eksempel 8: Slicing
bogstaver = ['a', 'b', 'c', 'd', 'e', 'f']

print('Hele listen:')
print(bogstaver)

print('\nElement 1-3:')
print(bogstaver[1:4])  # Husk: slut er IKKE inkluderet!

print('\nFørste 3 elementer:')
print(bogstaver[:3])

print('\nFra element 2 til slut:')
print(bogstaver[2:])
````
