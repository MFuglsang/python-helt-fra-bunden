## Datatyper (Data Types): Hvilken slags information gemmer vi?

Forestil dig, at du har forskellige typer af legetøj: nogle er tal, nogle er bogstaver, og nogle er billeder. På samme måde kan computere gemme forskellige slags information, som vi kalder **datatyper**.

Vi har allerede set én datatype, nemlig **tekst (strings)**. En tekststreng er en række bogstaver, tal eller symboler, som er sat sammen. Når du skriver dit navn, bruger du en tekststreng.

Python ved, at det er en tekststreng, når du sætter citationstegn omkring den, f.eks. `'Hej'` eller `"Verden"`.

Lad os prøve at lege mere med tekststrenge!

## Tal (Numbers): Hele tal og decimaltal

Udover tekststrenge kan computere også gemme tal. Der er primært to slags tal, vi skal kende til lige nu:

1.  **Hele tal (Integers)**: Det er tal uden decimaler, som f.eks. `1`, `10`, `-5`, `1000`. Tænk på dem som de tal, du tæller med.

2.  **Decimaltal (Floats)**: Det er tal med decimaler, som f.eks. `3.14`, `0.5`, `-2.75`. Her bruger vi et punktum i stedet for et komma.

Python kan udføre matematik med tal, ligesom en lommeregner! Når vi skal vise tal sammen med tekst, skal vi lige huske, at Python ofte har brug for at vide, at tallet skal behandles som tekst. Vi bruger funktionen `str()` til dette, men det vender vi tilbage til med en mere detaljeret forklaring senere!

Lad os prøve det.

```python
# Her er eksempler på hele tal (integers) og decimaltal (floats).
alder = 12
pris = 99.95
antal_elever = 28

# Vi kan også regne med dem.
sum_alder_og_elever = alder + antal_elever
# Husk at konvertere tal til tekst med str() når vi sætter dem sammen med tekst ved hjælp af '+'
print("Summen af alder og antal elever er: " + str(sum_alder_og_elever))

# Øvelse: Opret to variabler, en med dit yndlingstal (et helt tal) og en med din højde i meter (et decimaltal).
# Prøv derefter at gange dem sammen og print resultatet med en forklarende tekst, hvor du bruger '+' og str() til at sætte det hele sammen.
```

## Basis Matematik med Tal

Python er en fantastisk lommeregner! Vi kan udføre alle de almindelige matematiske operationer med vores `int` (hele tal) og `float` (decimaltal).

Her er de mest almindelige matematiske symboler, du skal kende:

*   `+`: Addition (lægger sammen)
*   `-`: Subtraktion (trækker fra)
*   `*`: Multiplikation (gange)
*   `/`: Division (dividerer)
*   `%`: Modulus (finder resten ved division)
*   `**`: Potens (opløfter i potens, f.eks. 2**3 er 2 ganget med sig selv 3 gange = 8)

Lad os prøve at regne lidt!

```python
# Eksempler på matematiske operationer

tal_a = 10
tal_b = 3
decimaltal_c = 5.0
decimaltal_d = 2.5

# Addition
sum_af_tal = tal_a + tal_b
print(f"Tal_a + tal_b = {sum_af_tal}") # Output: 13

# Subtraktion
forskel = tal_a - tal_b
print(f"Tal_a - tal_b = {forskel}") # Output: 7

# Multiplikation
produkt = tal_a * tal_b
print(f"Tal_a * tal_b = {produkt}") # Output: 30

# Division
kvotient = tal_a / tal_b
print(f"Tal_a / tal_b = {kvotient}") # Output: 3.333...

# Matematik med decimaltal
sum_decimal = decimaltal_c + decimaltal_d
print(f"Decimaltal_c + decimaltal_d = {sum_decimal}") # Output: 7.5

produkt_decimal = decimaltal_c * decimaltal_d
print(f"Decimaltal_c * decimaltal_d = {produkt_decimal}") # Output: 12.5

# Øvelse: Opret to nye variabler, en med et helt tal og en med et decimaltal.
# Udfør mindst tre forskellige matematiske operationer (f.eks. plus, gange, dividere) med disse to tal.
# Print resultatet af hver operation med en forklarende tekst, f.eks. "Summen er: [resultat]".
# Husk at bruge f-strings eller '+' og str() til at sætte teksten sammen!
```

## Brugerinput (User Input): Tal med computeren!

Indtil nu har vi fortalt computeren, hvad den skulle gøre med faste værdier. Men hvad nu hvis vi gerne vil have computeren til at spørge OS om noget?

Det kan vi gøre med funktionen `input()`. Når Python møder `input()`, stopper programmet og venter på, at brugeren (dig!) skriver noget og trykker Enter. Det, du skriver, bliver så gemt som en tekststreng.

Lad os prøve at få computeren til at spørge dig om dit navn!

```python
# Her bruger vi input() til at spørge om navnet.
# Teksten inden i parenteserne er den besked, brugeren ser.
dit_navn = input("Hvad er dit navn? ")

# Nu kan vi bruge det indtastede navn til at hilse på brugeren.
# Vi bruger '+' til at sætte teksten sammen.
print("Hej, " + dit_navn + "! Velkommen til Python.")

# Øvelse: Prøv at ændre beskeden i input() funktionen til noget andet.
# Få derefter computeren til at spørge om din yndlingsfarve og print en sætning med farven ved at bruge '+' til at sætte den sammen.
```

## Input og Tal: Pas på datatypen!

Da du brugte `input()` til at spørge om dit navn, fik Python en **tekststreng (string)** tilbage. Selv hvis du taster et tal, vil `input()` stadig se det som tekst!

Hvis vi vil regne med det tal, bruger vi funktionerne `int()` for hele tal (integers) eller `float()` for decimaltal. De 'omdanner' teksten til et rigtigt tal, som Python kan regne med.

Lad os prøve at lave en lille lommeregner!

```python
# Her spørger vi om to tal og omformer dem til integers.
tal1_tekst = input("Indtast det første tal: ")
tal1 = int(tal1_tekst)

tal2_tekst = input("Indtast det andet tal: ")
tal2 = int(tal2_tekst)

# Nu kan vi regne med dem!
sum_af_tal = tal1 + tal2
produkt_af_tal = tal1 * tal2

# Husk at konvertere tal til tekst med str() når vi sætter dem sammen med tekst ved hjælp af '+'
print("Summen af " + str(tal1) + " og " + str(tal2) + " er: " + str(sum_af_tal))
print("Produktet af " + str(tal1) + " og " + str(tal2) + " er: " + str(produkt_af_tal))

# Øvelse: Prøv at ændre koden, så den spørger om din alder og din yndlingstal.
# Beregn derefter, hvor gammel du vil være om 10 år, og print resultatet ved at sætte det sammen med '+' og str().
```

## Konvertering af Datatyper (Type Conversion)

Som vi så med `input()`, gemmer Python automatisk det, du skriver, som en tekststreng. Men hvad hvis vi har et tal, der er gemt som tekst, og vi gerne vil regne med det? Eller omvendt, hvis vi har et tal, vi gerne vil vise som tekst sammen med anden tekst?

Det kan vi gøre ved at 'konvertere' datatypen. Vi bruger specifikke funktioner til dette:

*   `int()`: Konverterer til et **helt tal** (integer).
*   `float()`: Konverterer til et **decimaltal** (float).
*   `str()`: Konverterer til en **tekststreng** (string).

Det er ligesom at ændre et stykke legetøj fra én kategori til en anden, så du kan bruge det til noget andet!

```python
# Eksempel på konvertering

alder_som_tekst = "13"
# Husk at konvertere typen() output til en string for at kunne sætte den sammen med tekst ved hjælp af '+'
print("'alder_som_tekst' er af typen: " + str(type(alder_som_tekst)))

alder_som_tal = int(alder_som_tekst)
print("'alder_som_tal' er af typen: " + str(type(alder_som_tal)))

# Nu kan vi regne med det!
fremtidig_alder = alder_som_tal + 5
# Husk at konvertere tal til tekst med str() når vi sætter dem sammen med tekst ved hjælp af '+'
print("Om fem år vil jeg være: " + str(fremtidig_alder) + " år.")

# Vi kan også konvertere et tal tilbage til tekst, hvis vi skal kombinere det med anden tekst.
temperatur = 21.5
tekst_om_temperatur = "Dagens temperatur er " + str(temperatur) + " grader."
print(tekst_om_temperatur)

# Øvelse: Opret en variabel med et decimaltal, f.eks. din vægt (f.eks. '65.3').
# Konverter dette decimaltal til et helt tal, og print begge variabler med en forklarende tekst og deres type.
# Brug '+' og str() til at sætte teksten sammen.
# Hvad sker der med decimalerne, når du konverterer til int()?
```

## Sæt tekst og tal sammen med '+' (String Concatenation)

Vi har allerede set, hvordan `print()` funktionen kan vise os information. Og vi har brugt `+` til at sætte tekststrenge sammen. Men hvad nu hvis vi vil sætte tekst og *tal* sammen?

Python kan ikke automatisk sætte et tal direkte sammen med en tekststreng ved hjælp af `+`. Du får en fejl, hvis du prøver direkte! Prøv selv: `print("Mit tal er: " + 5)` (denne kode vil give en fejl, hvis du kører den).

Vi skal fortælle Python, at tallet skal opføre sig som tekst, før vi kan sætte det sammen. Det gør vi med funktionen `str()`, som konverterer et tal til en tekststreng.

Det er ligesom at pakke et tal ind i en "tekst-indpakning", så det kan sendes med i en sætning.

Lad os prøve det!

```python
# Opret nogle variabler
navn = "Karl"
alder = 12
favorit_farve = "grøn"
favorit_tal = 10

# Brug '+' og str() til at lave en flot sætning
# Husk at konvertere tal til tekst med str()
print("Hej, jeg hedder " + navn + ", og jeg er " + str(alder) + " år gammel. Min yndlingsfarve er " + favorit_farve + ", og mit lykketal er " + str(favorit_tal) + "!")

# Øvelse: Opret dine egne variabler med dit navn, alder, yndlingsdyr og hvor mange kæledyr du har.
# Brug derefter '+' og str() til at printe en sætning, der fortæller om dig selv og dine kæledyr.
# F.eks: "Jeg er " + dit_navn + ", jeg er " + str(din_alder) + " år gammel, og mit yndlingsdyr er " + yndlingsdyr + ". Jeg har " + str(antal_kæledyr) + " kæledyr."
```

## F-strings: En nemmere måde at kombinere tekst og variabler på!

Nu har vi set, hvordan vi kan sætte tekst og tal sammen ved hjælp af `+` og funktionen `str()`. Det virker, men det kan godt blive lidt langt og besværligt at læse, især når man har mange variabler, der skal med.

Her kommer **f-strings** (eller *format-strenge*) ind i billedet! En f-string gør det super nemt og overskueligt at sætte variabler ind i en sætning. Du starter strengen med et `f` lige før citationstegnene (f.eks. `f"Denne tekst..."`), og så kan du skrive variabelnavne direkte inde i sætningen ved at omgive dem med krøllede parenteser `{}`.

Python tager sig automatisk af at konvertere tal til tekst, så de kan vises sammen med resten af sætningen. Det er en meget mere elegant måde at gøre det på!

Lad os prøve det!

```python
# Opret nogle variabler
mit_navn = "Anna"
min_alder = 13
min_yndlingsfarve = "blå"
mit_lykketal = 7

# Brug en f-string til at lave en flot sætning
print(f"Hej, jeg hedder {mit_navn}, og jeg er {min_alder} år gammel. Min yndlingsfarve er {min_yndlingsfarve}, og mit lykketal er {mit_lykketal}!")

# Øvelse: Opret dine egne variabler med dit navn, alder, yndlingsdyr og hvor mange kæledyr du har.
# Brug derefter en f-string til at printe en sætning, der fortæller om dig selv og dine kæledyr.
# F.eks: "Jeg er [Dit navn], jeg er [din alder] år gammel, og mit yndlingsdyr er [yndlingsdyr]. Jeg har [antal kæledyr] kæledyr."
# Sammenlign med den tidligere øvelse med '+', og se hvor meget nemmere f-strings er!
```

## Valutaomregner: DKK til Euro (med Decimaltal!)

Nu hvor vi kan modtage input og regne med decimaltal, kan vi lave en lille valutaomregner. Vi skal have brugeren til at indtaste et beløb i danske kroner (DKK), og så skal vi omregne det til Euro (EUR).

Vi antager, at kursen er 1 EUR = 7.46 DKK.

Husk at bruge `float()` til at konvertere brugerens input til et decimaltal, så du kan regne med det.

```python
# Øvelse: Opret en valutaomregner fra DKK til EUR.
# 1. Spørg brugeren om et beløb i DKK (f.eks. 'Hvor mange DKK vil du omregne? ').
# 2. Konverter inputtet til et decimaltal (float).
# 3. Definer vekselkursen (f.eks. `euro_kurs = 7.46`).
# 4. Beregn beløbet i EUR.
# 5. Print resultatet i en sætning, f.eks. 'X DKK svarer til Y Euro.'
# Brug f-strings til at gøre udskriften pæn og nem at læse.

# Dit kode her:
```

```python
# Løsning til Valutaomregner: DKK til EUR

dkk_amount_str = input("Hvor mange DKK vil du omregne? ")
dkk_amount = float(dkk_amount_str)

euro_kurs = 7.46

eur_amount = dkk_amount / euro_kurs

print(f"{dkk_amount} DKK svarer til {eur_amount:.2f} Euro.")
```

## Celsius til Fahrenheit Omregner!

Nu skal vi prøve at lave en lille omregner, der konverterer temperatur fra Celsius til Fahrenheit. Vi skal bede brugeren om at indtaste en temperatur i Celsius, og så skal vi udregne, hvad det svarer til i Fahrenheit.

Formlen for at omregne Celsius til Fahrenheit er:
`Fahrenheit = (Celsius * 9/5) + 32`

Husk at bruge `float()` til at konvertere brugerens input, så du kan regne med decimaltal.

```python
# Øvelse: Opret en omregner fra Celsius til Fahrenheit.
# 1. Spørg brugeren om en temperatur i Celsius (f.eks. 'Indtast temperatur i Celsius: ').
# 2. Konverter inputtet til et decimaltal (float).
# 3. Beregn temperaturen i Fahrenheit ved hjælp af formlen: `Fahrenheit = (Celsius * 9/5) + 32`.
# 4. Print resultatet i en sætning, f.eks. 'X grader Celsius svarer til Y grader Fahrenheit.'
# Brug f-strings til at gøre udskriften pæn og nem at læse.

# Dit kode her:
```

```python
# Løsning til Celsius til Fahrenheit Omregner

celsius_str = input("Indtast temperatur i Celsius: ")
celsius = float(celsius_str)

fahrenheit = (celsius * 9/5) + 32

print(f"{celsius} grader Celsius svarer til {fahrenheit:.2f} grader Fahrenheit.")
```

## Oversigt over Vigtige Termer til Undervisning

For at læreren kan formidle materialet bedst muligt, er her en opsummering af de centrale Python-begreber, som er introduceret i denne arbejdsbog, sammen med en kort forklaring:

*   **Variabel (Variable):** En "boks" i computerens hukommelse, der gemmer en værdi eller et stykke information. Man giver den et navn, så man kan hente informationen frem igen. *Analogi: En navngiven skuffe med indhold.*

*   **Datatyper (Data Types):** Forskellige slags information, som en computer kan håndtere. De primære datatyper introduceret her er tekst og tal.

    *   **Tekststreng (String / `str`):** En række af tegn (bogstaver, tal, symboler) omgivet af citationstegn (enkelt `' '` eller dobbelt `" "`). Bruges til navne, sætninger osv. *Analogi: Bogstaver og ord.*

    *   **Heltal (Integer / `int`):** Et tal uden decimaler. Kan være positive, negative eller nul. *Analogi: Tal, du tæller med.*

    *   **Decimaltal (Float / `float`):** Et tal med decimaler. Et punktum (`.`) bruges som decimalseparator. *Analogi: Målinger eller priser med komma.*

*   **Operatorer (Operators):** Symboler, der udfører handlinger på værdier (f.eks. matematiske operationer).

    *   `+` (Addition/Sammenkædning): Lægger tal sammen eller "kæder" tekststrenge sammen.
    *   `-` (Subtraktion): Trækker tal fra hinanden.
    *   `*` (Multiplikation): Ganger tal.
    *   `/` (Division): Dividerer tal.

*   **Funktion (Function):** En samling af kode, der udfører en specifik opgave. Vi har brugt flere indbyggede funktioner:

    *   `print()`: Viser information på skærmen.
    *   `input()`: Bedømmer brugeren om at indtaste information.
    *   `int()`: Konverterer en værdi til et heltal.
    *   `float()`: Konverterer en værdi til et decimaltal.
    *   `str()`: Konverterer en værdi til en tekststreng.

*   **String Konkatenation (String Concatenation):** Processen med at sætte to eller flere tekststrenge sammen ved hjælp af `+` operatoren. Kræver at tal konverteres til tekst med `str()` først.

*   **F-strings (Formatted String Literals):** En moderne og nem måde at bygge tekststrenge på i Python, hvor man kan indsætte variabler direkte ind i strengen ved at omgive dem med krøllede parenteser `{}`. Python håndterer automatisk konvertering fra tal til tekst.

Disse termer udgør fundamentet for forståelsen af de indledende programmeringskoncepter i Python.
