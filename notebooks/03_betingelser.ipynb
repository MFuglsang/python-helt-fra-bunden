## Introduktion til Betingelser

I programmering giver betingelser (eller kontrolstrukturer) os mulighed for at udføre forskellig kode baseret på, om en bestemt betingelse er sand eller falsk. Dette er fundamentalt for at skabe dynamiske og intelligente programmer.

## Introduktion til Booleans (True/False)

Inden vi dykker dybere ned i betingelser, er det vigtigt at forstå **Booleans**. En Boolean er en datatype, der kun kan have to værdier: `True` (sand) eller `False` (falsk). Disse værdier er fundamentale for at programmet kan træffe beslutninger.

Når vi udfører en sammenligning i programmering, stiller vi grundlæggende et spørgsmål, der kun kan besvares med 'sandt' eller 'falsk'. For eksempel:

*   `er_vejret_godt == True` (Er vejret godt, sandt?)
*   `alder >= 18` (Er alderen større end eller lig med 18?)
*   `temperatur < 0` (Er temperaturen mindre end nul?)

Disse spørgsmål har kun to mulige udfald: enten er betingelsen opfyldt (sandt), eller også er den ikke opfyldt (falsk). Python og de fleste andre programmeringssprog repræsenterer disse to udfald med `True` og `False` datatyperne – altså Booleans.

Resultatet af en sammenligning er derfor altid en Boolean, fordi sammenligningens natur er at evaluere en tilstand og returnere, hvorvidt denne tilstand er sand eller falsk.

### Sammenligningsoperatorer

Udover logiske operatorer er **sammenligningsoperatorer** afgørende for at skabe betingelser. Disse operatorer sammenligner to værdier og returnerer et Boolean-resultat (`True` eller `False`). Her er de mest almindelige:

*   `==` (er lig med)
*   `!=` (er ikke lig med)
*   `>` (er større end)
*   `<` (er mindre end)
*   `>=` (er større end eller lig med)
*   `<=` (er mindre end eller lig med)

```python
# Eksempler på sammenligningsoperatorer

a = 10
b = 5

print(f"a == b: {a == b}") # Er a lig med b?
print(f"a != b: {a != b}") # Er a ikke lig med b?
print(f"a > b: {a > b}")  # Er a større end b?
print(f"a < b: {a < b}")  # Er a mindre end b?
print(f"a >= b: {a >= b}") # Er a større end eller lig med b?
print(f"a <= b: {a <= b}") # Er a mindre end eller lig med b?
```

```python
# Eksempel på sammenligning der resulterer i en Boolean
print(f"Er 7 større end 9? {7 > 9}")
print(f"Er 10 lig med 10? {10 == 10}")

alder = 25
print(f"Er alder ({alder}) større end eller lig med 18? {alder >= 18}")

is_sunny = True
print(f"Er det solrigt? {is_sunny == True}") # Eller simpelthen just 'is_sunny'
```

I eksemplet ovenfor ser vi, hvordan boolske værdier oftest er resultatet af en sammenligning. Vi ser også et tilfælde, hvor vi tildeler `True` direkte til en variabel. Når vi spørger, om `alder >= 18`, evalueres udtrykket til enten `True` eller `False` baseret på `alder`s nuværende værdi. Disse `True`/`False`-resultater er præcis, hvad `if/else`-strukturer bruger til at afgøre, hvilken kodeblok der skal udføres.

## Logiske Operatorer (and, or, not)

Logiske operatorer bruges til at kombinere eller modificere betingelser. De er afgørende for at skabe komplekse beslutningsstrukturer i dine programmer. Python har tre grundlæggende logiske operatorer: `and`, `or`, og `not`.

### `and` Operatoren
`and` returnerer `True`, hvis **begge** betingelser er sande. Hvis bare én betingelse er falsk, returneres `False`.

### `or` Operatoren
`or` returnerer `True`, hvis **mindst én** af betingelserne er sand. Den returnerer kun `False`, hvis begge betingelser er falske.

### `not` Operatoren
`not` er en inverteringsoperator; den returnerer det modsatte af betingelsens sandhedsværdi. Hvis en betingelse er `True`, gør `not` den til `False`, og omvendt.

```python
# Eksempler på logiske operatorer

alder = 25
har_kørekort = True
har_bil = False

print(f"Alderen er 25 og har kørekort: {alder >= 18 and har_kørekort}") # True and True -> True
print(f"Alderen er 25 og har bil: {alder >= 18 and har_bil}") # True and False -> False

print(f"Har kørekort eller har bil: {har_kørekort or har_bil}") # True or False -> True
print(f"Har ikke kørekort og har ikke bil: {not har_kørekort and not har_bil}") # False and True -> False
```

### if-udsagn
Et `if`-udsagn udfører en blok kode, hvis en betingelse er sand.

I dette enkle `if`-udsagn tjekker vi værdien af variablen `alder`. Hvis `alder` er 18 eller derover (`alder >= 18`), vil sætningen inden i `if`-blokken (`print("Du er gammel nok til at stemme.")`) blive udført. Hvis betingelsen er falsk (f.eks. hvis `alder` er 17), vil koden inde i `if`-blokken blive sprunget over, og programmet fortsætter efter `if`-blokken.

```python
alder = 18

if alder >= 18:
    print("Du er gammel nok til at stemme.")
```

### if-else-udsagn
Et `if-else`-udsagn udfører én blok kode, hvis betingelsen er sand, og en anden blok kode, hvis betingelsen er falsk.

Her udvides funktionaliteten med en `else`-blok. Programmet evaluerer først `if`-betingelsen (`temperatur > 30`).
- Hvis betingelsen er sand, udføres koden i `if`-blokken.
- Hvis betingelsen er falsk, udføres koden i `else`-blokken i stedet. Kun én af blokkene vil blive udført.

```python
temperatur = 25

if temperatur > 30:
    print("Det er en varm dag!")
else:
    print("Temperaturen er behagelig.")
```

### if-elif-else-udsagn
`if-elif-else`-udsagn giver os mulighed for at tjekke flere betingelser i rækkefølge.

Når vi har flere mulige betingelser at tjekke, er `if-elif-else`-strukturen ideel. Programmet tjekker betingelserne i rækkefølge:
1. Først tjekkes `if score >= 90`.
2. Hvis den er falsk, tjekkes den næste `elif score >= 80`.
3. Hvis den også er falsk, tjekkes den næste `elif score >= 70`.
4. Hvis alle `if` og `elif` betingelser er falske, udføres koden i `else`-blokken. Så snart en betingelse er sand, udføres den tilhørende kodeblok, og resten af `elif`- og `else`-blokkene springes over.

```python
score = 85

if score >= 90:
    print("Karakter: 12")
elif score >= 80:
    print("Karakter: 10")
elif score >= 70:
    print("Karakter: 7")
else:
    print("Karakter: 4")
```

### Indlejrede Betingelser
Du kan også have betingelser inden i andre betingelser (indlejrede betingelser).

Indlejrede betingelser betyder, at du placerer et `if`- eller `if-else`-udsagn inde i en anden `if`- eller `else`-blok. Dette giver dig mulighed for at lave mere komplekse logikker, hvor en betingelse skal være opfyldt, før en anden betingelse overhovedet tjekkes.

I dette eksempel tjekkes det først, om `bruger_logget_ind` er sand. Hvis den er det, går programmet ind i den ydre `if`-blok og tjekker derefter, om `admin_bruger` også er sand. Hvis `bruger_logget_ind` er falsk, springes hele den indre logik over, og kun `else`-blokken for den ydre `if` udføres.

```python
bruger_logget_ind = True
admin_bruger = False

if bruger_logget_ind:
    print("Velkommen, bruger!")
    if admin_bruger:
        print("Du har administratorrettigheder.")
    else:
        print("Du er en standardbruger.")
else:
    print("Du skal logge ind for at fortsætte.")
```

### Ny Øvelse: Butikkens Rabat

En butik tilbyder forskellige rabatter baseret på det samlede købsbeløb:
- Hvis det samlede købsbeløb er **over 1000 DKK**, gives der **20% rabat**.
- Hvis det samlede købsbeløb er **over 500 DKK** (men ikke over 1000 DKK), gives der **10% rabat**.
- Ellers gives der **ingen rabat**.

Opret et program, der:
1.  Spørger brugeren om det samlede købsbeløb (`input()` og husk at konvertere til `float()`).
2.  Beregner den endelige pris efter rabat.
3.  Udskriver den oprindelige pris, den givne rabatprocent, og den endelige pris.

### Løsning til 'Butikkens Rabat'

Her er en mulig løsning på opgaven. Forsøg at løse den selv, før du kigger på løsningen!

```python
original_pris_str = input("Indtast det samlede købsbeløb i DKK: ")
original_pris = float(original_pris_str)

rabat_procent = 0

if original_pris > 1000:
    rabat_procent = 20
elif original_pris > 500:
    rabat_procent = 10
else:
    rabat_procent = 0

final_pris = original_pris * (1 - rabat_procent / 100)

print(f"\nOprindelig pris: {original_pris:.2f} DKK")
print(f"Anvendt rabat: {rabat_procent}% ")
print(f"Endelig pris: {final_pris:.2f} DKK")
```
