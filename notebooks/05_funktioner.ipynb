## Hvad er en Funktion?

En funktion er som en opskrift i kogebogen.

### Eksempel fra virkeligheden:

**Uden funktion:**
```
Dag 1:
    Hvælve mel i skål
    Hvælve sukker i skål
    Bland det sammen
    Lav pandekagetej

Dag 2:
    Hvælve mel i skål
    Hvælve sukker i skål
    Bland det sammen
    Lav pandekagetej
```

**Med funktion:**
```
lav_pandekagetej():
    Hvælve mel i skål
    Hvælve sukker i skål
    Bland det sammen
    Lav pandekagetej

Dag 1: lav_pandekagetej()
Dag 2: lav_pandekagetej()
```

Meget mere praktisk!

## Din Første Funktion

Her er syntaksen for at skrive en funktion:

```python
def navn_på_funktion():
    # Koden der skal køre
    print('Hej fra funktionen!')
```

### Vigtige dele:
- `def` - betyder "definer" en ny funktion
- `navn_på_funktion` - det navn du vælger for funktionen
- `()` - tomt hvis funktionen ikke skal have input (vi kommer til parameters senere)
- `:` - kolon efter funktionen (vigtig!)
- Koden under er *indrykket* (rykket ind) - det betyder det er "inde i" funktionen

### Vigtig regel:
Når du *definerer* en funktion, kører den IKKE automatisk!
Du skal *kalde* funktionen med `navn_på_funktion()`

### Lad os Prøve det!

````python
# Eksempel 1: Definer en simpel funktion
def siger_hej():
    print('Hej, verden!')

# Nu kalder vi funktionen - her kører den!
siger_hej()
siger_hej()
siger_hej()
````

**Hvad sker der?**

Vi definerede funktionen `siger_hej()` - men intet skete endnu.

Derefter kaldte vi funktionen 3 gange med `siger_hej()` - og "Hej, verden!" blev udskrevet 3 gange.

Det var meget lettere end at skrive `print('Hej, verden!')` tre gange!

## Funktioner med Indput - Parameters

En funktion bliver meget mere kraftfuld hvis den kan tå indput!

En "parameter" er som en variabel der fungerer som indput til funktionen.

````python
# Eksempel 2: Funktion med en parameter
def siger_hej_til(navn):
    print('Hej ' + navn + '!')

# Nu kalder vi funktionen med forskellige navne
siger_hej_til('Maria')
siger_hej_til('Anders')
siger_hej_til('Python')
````

**Hvad sker her?**

- Vi definerede funktionen `siger_hej_til(navn)` - `navn` er en parameter
- Hver gang vi kalder funktionen, giver vi den en værdi
- `siger_hej_til('Maria')` - her bliver `navn = 'Maria'`
- `siger_hej_til('Anders')` - her bliver `navn = 'Anders'`

Samme funktion, forskellige resultater!

## Funktioner med Output - return

Nogle gange vil du at en funktion skal give dig en værdi tilbage. Du bruger `return` til det!

`return` betyder: "Send denne værdi tilbage fra funktionen"

````python
# Eksempel 3: Funktion der returner en værdi
def ganger_med_to(tal):
    resultat = tal * 2
    return resultat

# Nu kalder vi funktionen og gemmer resultatet
svar1 = ganger_med_to(5)
svar2 = ganger_med_to(10)

print('5 * 2 = ' + str(svar1))
print('10 * 2 = ' + str(svar2))
````

**Hvad sker her?**

- Funktionen `ganger_med_to(tal)` tager et tal som indput
- Den ganger det med 2
- `return resultat` sender resultatet tilbage til den der kaldte funktionen
- Vi gemmer resultatet i en ny variabel: `svar1 = ganger_med_to(5)`
- Nu kan vi bruge `svar1` senere i programmet!

## Funktioner med Multiple Parameters

En funktion kan have mere end én parameter!

````python
# Eksempel 4: Funktion med 2 parameters
def lægger_sammen(tal1, tal2):
    sum_af_tal = tal1 + tal2
    return sum_af_tal

# Kald funktionen med 2 værdier
resultat = lægger_sammen(10, 25)
print('10 + 25 = ' + str(resultat))

# Du kan også bruge den direkte i print
print('7 + 3 = ' + str(lægger_sammen(7, 3)))
````

**Hvad sker her?**

- Funktionen har 2 parameters: `tal1` og `tal2`
- Når vi kalder den, giver vi 2 værdier: `lægger_sammen(10, 25)`
- Første værdi (10) bliver `tal1`
- Anden værdi (25) bliver `tal2`

## Funktioner med Løkker og Betingelser

Du kan kombinere alt hvad du har lært! Funktioner kan indeholde løkker og if-statements!

````python
# Eksempel 5: Tjek hvis tal er lige eller ulige
def er_tal_lige(tal):
    if tal % 2 == 0:
        return 'lige'
    else:
        return 'ulige'

print('5 er ' + er_tal_lige(5))
print('8 er ' + er_tal_lige(8))
print('100 er ' + er_tal_lige(100))
````

**Hvad sker her?**

- Vi har en `if` statement inde i funktionen
- Hvis tallet er lige, returnerer vi 'lige'
- Ellers returnerer vi 'ulige'
- Vi bruger `return` til at sende forskellige værdier tilbage!

## Funktioner der Printer UND Returnerer

En funktion kan både printe noget og returnere noget!

````python
# Eksempel 6: Funktion der printer og returnerer
def sum_med_besked(tal1, tal2):
    sum_af_tal = tal1 + tal2
    print('Jeg lægger ' + str(tal1) + ' og ' + str(tal2) + ' sammen')
    return sum_af_tal

resultat = sum_med_besked(15, 20)
print('Resultatet er ' + str(resultat))
````

**Hvad sker her?**

- Funktionen printer en besked
- Og returnerer resultatet
- Vi kan bruge resultatet efter at funktionen er kørt!

## Funktioner der Kalder Andre Funktioner

Du kan have funktioner der kalder andre funktioner du har skrevet!

````python
# Eksempel 7: Funktioner der kalder hinanden
def udskriv_strek():
    print('---')

def udskriv_header(tekst):
    udskriv_strek()  # Kalder anden funktion!
    print(tekst)
    udskriv_strek()  # Kalder anden funktion igen!

# Brug header-funktionen
udskriv_header('Min Overskrift')
udskriv_header('Anden Overskrift')
````

**Hvad sker her?**

- `udskriv_strek()` er en simpel funktion der printer en streg
- `udskriv_header(tekst)` kalder `udskriv_strek()` to gange
- Funktioner kan kalde andre funktioner!

## Fælles Fejl - Pas På!

### Fejl 1: Glemmer Indentation
```python
def min_funktion():
print('Hej')  # FEJL - ikke rykket ind!
```
**Løsning:** Ryk funktionen-indholdet ind

### Fejl 2: Glemmer Kolon
```python
def min_funktion()  # FEJL - mangler kolon!
    print('Hej')
```
**Løsning:** Læg altid `:` efter funktionens navn

### Fejl 3: Kalder Funktionen Uden Parenteser
```python
min_funktion  # FEJL - intet sker!
```
**Løsning:** Husk parenteser: `min_funktion()`

### Fejl 4: Bruger return Værdi uden at Gemme Det
```python
def ganger_med_to(tal):
    return tal * 2

ganger_med_to(5)  # Resultatet "forsvinder"!
```
**Løsning:** Gem resultatet i en variabel
```python
resultat = ganger_med_to(5)  # Nu har vi resultatet!
```

### Fejl 5: Kalder Funktionen Før du Har Defineret Den
```python
siger_hej()  # FEJL - funktionen eksisterer ikke endnu!

def siger_hej():
    print('Hej!')
```
**Løsning:** Definer funktionen først, kald den derefter
