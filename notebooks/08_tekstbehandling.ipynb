## Hvorfor er tekstbehandling vigtig?

Tænk på ting du gør med tekst:
- Fjerne mellemrum fra en bruger-input
- Konvertere tekst til STORE BOGSTAVER eller små bogstaver
- Dele en sætning op i ord
- Søge efter ord i en tekst
- Tælle bokstaver eller ord
- Validere email-adresser

Alt dette kan du nu gøre med **string metoder**!

**Vigtig huskeseddel:** Strings i Python kan **ikke ændres på plads** (de er "immutable"). Når du bruger en string-metode, returnerer den en **ny string**:

````python
# Vigtig: Strings kan ikke ændres!
tekst = 'Hej'

# Du kan ikke gøre: tekst[0] = 'H'  (fejl!)
# I stedet skal du lave en ny string:
ny_tekst = tekst.upper()  # Laver ny string

print('Original:', tekst)
print('Ny version:', ny_tekst)
````

## Eksempel 1: Konverter til STORE og små bogstaver

De mest almindelige string-metoder handler om at ændre størrelse på bogstaver:

````python
# Eksempel 1: Bogstav-størrelse metoder
navn = 'Anna'

# Lav alt STORT
print(navn.upper())

# Lav alt småt
print(navn.lower())

# Lav kun første bogstav STORT
print(navn.capitalize())

# Lav første bogstav i hver ord STORT (titel-stil)
sætning = 'jeg holder af programmering'
print(sætning.title())
````

**Hvad skete der?**
- `.upper()` - alle bogstaver bliver STORE
- `.lower()` - alle bogstaver bliver små
- `.capitalize()` - kun første bogstav bliver STORT
- `.title()` - første bogstav i hver ord bliver STORT (god til titler)

Disse metoder returnerer altid en **ny string** - originalen ændres ikke.

## Eksempel 2: Erstat tekst med .replace()

Hvis du vil bytte nogle ord ud for andre ord, bruger du `.replace()`:

````python
# Eksempel 2: Erstat tekst
sætning = 'Jeg holder af hunde fordi hunde er søde'

# Erstat "hunde" med "katte"
ny_sætning = sætning.replace('hunde', 'katte')
print(ny_sætning)

# Du kan også specificere hvor mange gange
ny_sætning2 = sætning.replace('hunde', 'fugle', 1)  # Kun første forekomst
print(ny_sætning2)
````

**Hvad skete der?**
- `.replace('gammelt', 'nyt')` bytter alle forekomster
- `.replace('gammelt', 'nyt', 1)` bytter kun første forekomst
- Som altid returnerer det en ny string

## Eksempel 3: Fjern mellemrum med .strip()

Ofte får du input fra brugere der har "ekstra" mellemrum på siden. `.strip()` fjerner det:

````python
# Eksempel 3: Fjern mellemrum
tekst = '   Hej verden   '

print('Original: "' + tekst + '"')
print('Efter .strip(): "' + tekst.strip() + '"')

# Der er også:
# .lstrip() - fjerner mellemrum fra venstre side
# .rstrip() - fjerner mellemrum fra højre side
print('\nLstrip: "' + tekst.lstrip() + '"')
print('Rstrip: "' + tekst.rstrip() + '"')
````

**Hvad skete der?**
- `.strip()` fjerner mellemrum fra både start og slutning
- `.lstrip()` fjerner kun fra venstre (left)
- `.rstrip()` fjerner kun fra højre (right)

Dette er **super vigtig** når du modtager input fra brugere!

## Eksempel 4: Opdel tekst med .split()

Hvis du har en sætning og vil have hver ord som en **liste**, bruger du `.split()`:

````python
# Eksempel 4: Opdel tekst med .split()
sætning = 'Jeg holder af programmering og Python'

# Opdel på mellemrum (default)
ord_liste = sætning.split()
print(ord_liste)
print('Antal ord:', len(ord_liste))

# Opdel på noget andet
csv = 'Anna,Bert,Cecilia,David'
navn_liste = csv.split(',')
print('\nNavne:', navn_liste)
````

**Hvad skete der?**
- `.split()` uden argument opdeler på mellemrum
- `.split(',')` opdeler på kommaer
- Det returnerer altid en **liste**!

Dette er **mega nyttigt** for at behandle data der kommer fra brugere eller filer.

## Eksempel 5: Sæt tekst sammen med .join()

`.join()` er det modsatte af `.split()` - det sætter en liste af ord sammen til en sætning:

````python
# Eksempel 5: Sæt sammen med .join()
ord_liste = ['Jeg', 'holder', 'af', 'Python']

# Sæt sammen med mellemrum
sætning = ' '.join(ord_liste)
print(sætning)

# Sæt sammen med kommaer
navne = ['Anna', 'Bert', 'Cecilia']
csv = ', '.join(navne)
print(csv)

# Sæt sammen uden separator
bogstaver = ['H', 'e', 'j']
ord = ''.join(bogstaver)
print(ord)
````

**Hvad skete der?**
- `.join()` skal altid bruges på en string (separatoren)
- `' '.join(liste)` sætter sammen med mellemrum
- `','.join(liste)` sætter sammen med kommaer
- `''.join(liste)` sætter sammen uden noget

Bemærk: `' '.join(['a', 'b'])` ikke `['a', 'b'].join(' ')`!

## Eksempel 6: Søg efter tekst med "in"

Hvis du vil tjekke om et ord eller bogstav eksisterer i en tekst, bruger du `in`:

````python
# Eksempel 6: Søg efter tekst
sætning = 'Jeg holder af programmering'

# Tjek hvis ordet eksisterer
if 'programmering' in sætning:
    print('Ordet findes!')

# Tjek hvis ordet IKKE eksisterer
if 'hunde' not in sætning:
    print('Ordet findes ikke')

# Tjek efter bogstav
if 'Z' in sætning.upper():
    print('Bogstavet Z findes')
else:
    print('Bogstavet Z findes ikke')
````

**Hvad skete her?**
- `'ord' in tekst` returnerer True hvis ordet findes
- `'ord' not in tekst` returnerer True hvis ordet IKKE findes
- Det er **case-sensitive** - "Programmering" og "programmering" er forskellige!
- Brug `.lower()` eller `.upper()` hvis du ikke vil være case-sensitive

## Eksempel 7: Find position med .find()

Hvis du vil vide **hvor** i en tekst et ord eller bogstav befinder sig, bruger du `.find()`:

````python
# Eksempel 7: Find position
sætning = 'Jeg holder af programmering'

# Find hvor ordet "programmering" starter
position = sætning.find('programmering')
print('Ordet "programmering" starter på plads:', position)

# Find hvor bogstavet 'o' er
position2 = sætning.find('o')
print('Første "o" er på plads:', position2)

# Hvis ordet ikke findes, returnerer det -1
position3 = sætning.find('hunde')
print('Ordet "hunde" findes på plads:', position3)
````

**Hvad skete her?**
- `.find()` returnerer **indeks** (position) af første forekomst
- Hvis ordet ikke findes, returnerer det `-1` (ikke fejl!)
- Du kan bruge det til at tjekke: `if tekst.find('ord') != -1:` betyder ordet findes

## Eksempel 8: Tjek start og slutning

`.startswith()` og `.endswith()` fortæller om en tekst starter eller slutter med noget:

````python
# Eksempel 8: Start og slutning
email = 'anna@gmail.com'

# Tjek hvis emailen slutter med .com
if email.endswith('.com'):
    print('Gyldig .com email')

# Tjek hvis emailen starter med et bestemt navn
if email.startswith('anna'):
    print('Hej Anna!')

# Praktisk: Validering af fil-typer
fil = 'billede.jpg'
if fil.endswith('.jpg') or fil.endswith('.png'):
    print('Gyldig billedfil')
````

**Hvad skete her?**
- `.startswith('tekst')` returnerer True hvis teksten starter med det
- `.endswith('tekst')` returnerer True hvis teksten slutter med det
- Dette er **super nyttigt** til at validere email, fil-typer, osv.

## Eksempel 9: String indeksering og slicing

Strings virker næsten som lister - du kan få enkelte bogstaver eller dele af strings:

````python
# Eksempel 9: String indeksering
ord = 'Python'

# Get enkelte bogstaver
print('Første bogstav:', ord[0])
print('Andet bogstav:', ord[1])
print('Sidste bogstav:', ord[-1])

# Get dele af streng (slicing)
print('Første 3 bogstaver:', ord[0:3])
print('Fra bogstav 2 til slutning:', ord[2:])
print('Sidste 3 bogstaver:', ord[-3:])

# Løkke gennem bogstaver
print('\nHver bogstav:')
for bogstav in ord:
    print(bogstav)
````

**Hvad skete her?**
- Strings bruger samme indeks-syntaks som lister!
- `string[0]` er første bogstav
- `string[-1]` er sidste bogstav
- `string[1:4]` er bogstaver fra plads 1 til 3 (ikke til 4)
- Du kan løkke gennem bogstaver med `for bogstav in string:`

## Eksempel 10: Praktisk eksempel - Tæl bogstaver

Lad os lave noget praktisk - tæl hvor mange vocaler der er i en sætning:

````python
# Eksempel 10: Tæl vocaler
sætning = 'jeg holder af programmering'
vocaler = 'aeiouæøå'

tæller = 0
for bogstav in sætning:
    if bogstav in vocaler:
        tæller = tæller + 1

print('Antal vocaler:', tæller)

# Vi kunne også bruge .count()
antal_a = sætning.count('a')
antal_e = sætning.count('e')
print('Antal a:\'er:', antal_a)
print('Antal e:\'er:', antal_e)
````

**Hvad skete her?**
- Vi løkkede gennem hver bogstav i sætningen
- Vi tjekker hvis bogstavet er en vokal
- Vi tæller hvor mange vocaler der er i alt
- `.count()` er en anden metode der tæller forekomster

## Eksempel 11: Praktisk - Email validering

Lad os lave en simpel email-validering (ikke rigtig, men simpel):

````python
# Eksempel 11: Simpel email-validering
email = 'anna.jensen@gmail.com'

# Email skal have @ og ende med .com eller .dk osv.
if '@' in email and '.' in email:
    if email.endswith('.com') or email.endswith('.dk'):
        print('Email ser gyldig ud')
    else:
        print('Email skal ende med .com eller .dk')
else:
    print('Email skal have @ og punkt')
````

**Hvad skete her?**
- Vi kombinerede flere string-metoder for at validere en email
- Vi tjekker hvis `@` og `.` eksisterer
- Vi tjekker hvis den slutter med gyldigt domæne
- Dette er "toy validation" - rigtig validation er mere kompleks!

## Eksempel 12: Løkke + String Metoder

Lad os lave et eksempel der kombinerer løkker og string-metoder:

````python
# Eksempel 12: Find alle ord der starter med S
sætning = 'Superb stort spil skaber sjov og stimulation'
ord_liste = sætning.lower().split()  # Lav småt og opdel

print('Ord der starter med S:')
for ord in ord_liste:
    if ord.startswith('s'):
        print(ord)
````

**Hvad skete der?**
- Vi brugte `.lower()` til at gøre alt småt (så S og s er det samme)
- Vi brugte `.split()` til at få en liste af ord
- Vi løkkede gennem ord-listen
- Vi brugte `.startswith()` til at tjekke hvilket ord der starter med 's'

Se hvor kraftfuldt det bliver når du kombinerer ting!

## Fælles fejl 🐛

### Fejl 1: Glemmer at metoder returnerer ny string

**FORKERT:**
```python
tekst = 'Hej'
tekst.upper()  # Gør ingenting
print(tekst)  # Stadig "Hej" (småt)
```

**KORREKT:**
```python
tekst = 'Hej'
tekst = tekst.upper()  # Gem det nye resultat
print(tekst)  # Nu "HEJ"
```

**Forklaring:** String-metoder returnerer en ny string; de ændrer IKKE originalen.

---

### Fejl 2: Glemmer at .join() bruges på separator

**FORKERT:**
```python
liste = ['a', 'b', 'c']
tekst = liste.join(',')  # FEJL!
```

**KORREKT:**
```python
liste = ['a', 'b', 'c']
tekst = ','.join(liste)  # OK
```

**Forklaring:** Du bruger `.join()` på **separatoren** (kommaen), ikke på listen!

---

### Fejl 3: Blander .split() og .join()

**FORKERT:**
```python
tekst = 'a b c'
result = tekst.join(' ')  # Fejl - det er .split() du skal bruge!
```

**KORREKT:**
```python
tekst = 'a b c'
liste = tekst.split(' ')  # OK - fra tekst til liste
```

**Forklaring:**
- `.split()` → gør tekst til liste
- `.join()` → gør liste til tekst

---

### Fejl 4: Case-sensitivity problemer

**FORKERT:**
```python
sætning = 'Python er sjovt'
if 'PYTHON' in sætning:  # Fejl - 'PYTHON' != 'Python'
    print('Findes')
```

**KORREKT:**
```python
sætning = 'Python er sjovt'
if 'PYTHON'.lower() in sætning.lower():  # OK
    print('Findes')
```

**Forklaring:** Strings er case-sensitive. Lav begge til samme størrelse før sammenligning.

---

### Fejl 5: Glemmer at .find() returnerer -1 hvis ikke findes

**FORKERT:**
```python
tekst = 'Hej verden'
if tekst.find('xyz'):  # FEJL - returnerer -1, som er "sand"!
    print('Findes')
```

**KORREKT:**
```python
tekst = 'Hej verden'
if tekst.find('xyz') != -1:  # OK
    print('Findes')
# Eller bedre:
if 'xyz' in tekst:  # Mere læseligt
    print('Findes')
```

**Forklaring:** Brug `in` operator i stedet - det er mere læseligt.

---

### Fejl 6: Blander string-indeks med liste-operationer

**FORKERT:**
```python
tekst = 'Hej'
tekst.append('!')  # FEJL - strings har ikke .append()!
```

**KORREKT:**
```python
tekst = 'Hej'
tekst = tekst + '!'  # OK - string concatenation
```

**Forklaring:** Strings og lister har forskellige metoder. For strings bruger du `+` for at tilføje.
