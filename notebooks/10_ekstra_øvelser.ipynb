### Øvelse 4.5: Tegn en kasse
**Moduler:** 05 Funktioner, 04 Løkker

Lav en funktion `tegn_kasse(bredde, højde)` der tegner en kasse:
```

→
*****
*   *
*****
```

````python
# Løsning til 4.5
````

<details>
<summary>▼ Se løsning</summary>

```python
def tegn_kasse(bredde, højde):
    # Top
    print('*' * bredde)
    # Midt
    for i in range(højde - 2):
        print('*' + ' ' * (bredde - 2) + '*')
    # Bund
    print('*' * bredde)

tegn_kasse(5, 3)
```
</details>

### Øvelse 8.1: Mini-Quiz System
**Moduler:** 02 Variabler, 03 Betingelser, 04 Løkker, 05 Funktioner, 07 Dictionaries

Lav et quiz-system med spørgsmål gemme som dictionary:
```
spørgsmål = {
    'Hvad er 2+2?': '4',
    'Er Python et programmeringssprog?': 'ja'
}

Stillede begge spørgsmål, tæl rigtige svar

```

````python
# Løsning til 8.1
````

<details>

    'Hvad er 2+2?': '4',

for sp, svar in spørgsmål.items():
    print('Spørgsmål ' + str(i) + ': ' + sp)
    dit_svar = input('Dit svar: ')
    if dit_svar == svar:
        print('Rigtigt!')
        rigtige = rigtige + 1
    else:
        print('Forkert. Svaret er: ' + svar)
    i = i + 1

### Øvelse 8.2: Tekst-statistik
**Moduler:** 02 Variabler, 04 Løkker, 06 Lister, 07 Dictionaries, 08 Tekstbehandling

Analyser en tekst og report:
- Antal ord
- Antal tegn
- Hyppigste ord
- Mest brugte bogstav

```
tekst = 'Python er sjovt. Python er nemt.'
→
Ord: 6
Tegn: 33
Mest brugt ord: Python (2)
Mest brugt bogstav: a (3)
```

````python
# Løsning til 8.2
````

<details>
<summary>▼ Se løsning</summary>

```python
tekst = 'Python er sjovt. Python er nemt.'
ord_liste = tekst.lower().split()
print('Ord: ' + str(len(ord_liste)))

# Ret tegn (uden mellemrum og punktum)
rent_tekst = tekst.replace(' ', '').replace('.', '')
print('Tegn: ' + str(len(rent_tekst)))

# Mest brugt ord
ord_tæller = {}
for o in ord_liste:
    o_ren = o.replace('.', '')
    if o_ren in ord_tæller:
        ord_tæller[o_ren] = ord_tæller[o_ren] + 1
    else:
        ord_tæller[o_ren] = 1

mest_brugt = max(ord_tæller, key=ord_tæller.get)
print('Mest brugt ord: ' + mest_brugt + ' (' + str(ord_tæller[mest_brugt]) + ')')
```
</details>
