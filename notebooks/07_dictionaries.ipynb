# Modul 07: Ordbøger (Dictionaries) 📖

Nu er der kommet mere data i dine programmer! I de sidste moduler lærte du om **lister** - nu skal du lære om **ordbøger** (dictionaries), som er en anden og ofte smarter måde at organisere data på.

## Hvad er en ordbog (dictionary)?

Tænk på en rigtig ordbog:
- Du slår et **ord** op (nøgle)
- Du får **betydningen** frem (værdi)

**Eksempel:**
- Ord → Betydning
- "hund" → "Et husdyr som biver og æder kibble"
- "bil" → "Et køretøj med hjul"

En **dictionary** i Python virker på samme måde:
- Du har **nøgler** (keys) - som ordet i ordbogen
- Du har **værdier** (values) - som betydningen

**Vigtig forskel mellem liste og dictionary:**
- **Liste**: Bruger **nummer** (index) for at finde data → `min_liste[0]`, `min_liste[1]`
- **Dictionary**: Bruger **navn** (nøgle) for at finde data → `min_ordbog['navn']`, `min_ordbog['alder']`

Når du vil finde noget i et dictionary, skal du ikke huske hvilket nummer det er - du skal bare huske **navnet** på det!

## Hvornår Bruger Vi Dictionaries?

Dictionaries bruges overalt når du skal "slå noget op"! Her er eksempler:

- **Telefonbog**: Navn → Telefonnummer
- **Brugerdata**: Brugernavn → Email, Adresse, Alder
- **Lagersystem**: Produktkode → Pris, Antal på lager, Kategori
- **Karakterkort**: Elev-navn → Karakter
- **Spillerinfo**: Spiller-id → Navn, Score, Level
- **Ordbog**: Dansk-ord → Engelsk-oversættelse

**Lister vs Dictionaries:**
- Liste: "Giv mig element nummer 5"
- Dictionary: "Giv mig værdien for nøgle 'Anna'"

Dictionaries er **perfekte for lookup-operationer**!

## Eksempel 1: Opret en simpel dictionary

Her er den simpleste måde at lave en dictionary på - ved at give den nogle nøgle-værdi par med det samme:

````python
# Eksempel 1: Opret en dictionary med personer-oplysninger
person = {'navn': 'Anna', 'alder': 12, 'by': 'København'}

print(person)
````

**Hvad skete der?**
- Vi lavede en dictionary kaldet `person`
- Den har 3 nøgler: `'navn'`, `'alder'`, `'by'`
- Den har 3 værdier: `'Anna'`, `12`, `'København'`
- Hver nøgle er parret med sin værdi ved hjælp af `:` (kolon)
- Hele ordbogen bruger `{}`  krøllede paranteser (ikke `[]` som lister)

## Eksempel 2: Få værdi fra dictionary (adgang)

Hvordan finder du værdi på en nøgle? Du bruger `[]` lige som med lister - men med **navnet på nøglen** i stedet for et nummer:

````python
# Eksempel 2: Få værdier fra dictionary
person = {'navn': 'Anna', 'alder': 12, 'by': 'København'}

# Get værdier ved at skrive nøglen
print(person['navn'])
print(person['alder'])
print(person['by'])
````

**Hvad skete der?**
- `person['navn']` giver os værdien tilknyttet nøglen `'navn'` - altså `'Anna'`
- `person['alder']` giver os værdien `12`
- `person['by']` giver os værdien `'København'`

Det er meget smartere end at huske "det er på plads 0, 1, 2" - du kan bare huske navnene!

## Eksempel 3: Tilføj nye nøgle-værdi par

Du kan tilføje nye oplysninger til dictionaries, selv efter du har oprettet den:

````python
# Eksempel 3: Tilføj nye data til en dictionary
person = {'navn': 'Anna', 'alder': 12}

# Tilføj email
person['email'] = 'anna@mail.com'

# Tilføj favorit farve
person['farve'] = 'blå'

print(person)
````

**Hvad skete der?**
- Vi startede med 2 nøgle-værdi par
- Vi tilføjede `'email'` som en ny nøgle med værdi `'anna@mail.com'`
- Vi tilføjede `'farve'` som en ny nøgle med værdi `'blå'`
- Nu har dictionaries 4 ting i sig!

Det virker som at tilføje til en liste, men med navne i stedet for numre.

## Eksempel 4: Ændrer eksisterende værdier

Hvis du vil ændre noget der allerede er der, kan du bare assignere en ny værdi til nøglen:

````python
# Eksempel 4: Ændrer værdi i dictionary
person = {'navn': 'Anna', 'alder': 12, 'by': 'København'}

print('Før ændring:')
print(person)

# Anna bliver 13 år
person['alder'] = 13

# Anna flytter til Aarhus
person['by'] = 'Aarhus'

print('\nEfter ændring:')
print(person)
````

**Hvad skete der?**
- Vi brugte den samme syntaks som at tilføje nye værdier
- Men fordi `'alder'` og `'by'` **allerede eksisterer**, blev de ændret i stedet for at tilføje nye
- Hvis nøglen ikke eksisterer → tilføj ny
- Hvis nøglen eksisterer → ændrer værdi

## Eksempel 5: Slet data fra dictionary

Hvis du vil fjerne en nøgle-værdi par helt, kan du bruge `del`:

````python
# Eksempel 5: Slet data fra dictionary
person = {'navn': 'Anna', 'alder': 12, 'by': 'København', 'email': 'anna@mail.com'}

print('Før sletning:')
print(person)

# Slet email
del person['email']

print('\nEfter sletning af email:')
print(person)
````

**Hvad skete her?**
- `del person['email']` fjernede helt nøglen `'email'` og dens værdi
- Nu er der kun 3 ting tilbage i dictionaries

Bemærk: `del` virker også på lister - `del min_liste[0]` sletter element på plads 0.

## Eksempel 6: Løkke gennem en dictionary

Hvis du vil gøre noget med hver nøgle-værdi par, kan du bruge en `for`-løkke:

````python
# Eksempel 6: Løkke gennem alle nøgler i dictionary
person = {'navn': 'Anna', 'alder': 12, 'by': 'København'}

print('Alle nøgler:')
for noekle in person:
    print(noekle)
````

**Hvad skete der?**
- Vi løkkede gennem `person`
- Hver gang gav løkken os en **nøgle** (ikke værdi!)
- Vi fik: `'navn'`, `'alder'`, `'by'`

Bemærk: Når du løkker gennem en dictionary direkte, får du kun nøglerne.

## Eksempel 7: Løkke gennem nøgler OG værdier

Hvis du vil have både nøgler **og** værdier i løkken, bruger du `.items()` metoden:

````python
# Eksempel 7: Løkke gennem nøgler og værdier
person = {'navn': 'Anna', 'alder': 12, 'by': 'København'}

print('Alle nøgler og værdier:')
for noekle, vaerdi in person.items():
    print(noekle + ': ' + str(vaerdi))
````

**Hvad skete der?**
- `.items()` er en metode der giver os **både nøgler og værdier**
- Vi brugte to variabler i løkken: `noekle` og `vaerdi`
- Nu kan vi udskrive begge dele pent: `'navn: Anna'`, `'alder: 12'`, osv.

Dette er meget smarter for at vise dictionary-data på en pæn måde!

## Eksempel 8: Praktisk eksempel - Telefonbog

Lad os lave noget praktisk - en simpel telefonbog hvor du kan slå telefonnumre op på navn:

````python
# Eksempel 8: Telefonbog som dictionary
telefonbog = {
    'Anna': '40123456',
    'Bert': '40654321',
    'Cecilia': '40999999'
}

# Find Annas telefonnummer
print('Annas nummer: ' + telefonbog['Anna'])

# Tilføj en ny person
telefonbog['David'] = '40111111'

# Print hele telefonbogen pent
print('\nHel telefonbog:')
for navn, nummer in telefonbog.items():
    print(navn + ': ' + nummer)
````

**Hvad skete her?**
- Vi lavede en telefonbog hvor navne er nøgler og numre er værdier
- Vi kunne slå nummeret op på Annas navn
- Vi tilføjede David
- Vi printede hele telefonbogen på en pæn måde

Med en liste ville du skulle huske "David er på plads 3" - med dictionary kan du bare huske "David"!

## Eksempel 9: Karakterbog - Praktisk eksempel

Her er en anden praktisk brug - en karakterbog der gemmer karakterer for elever:

````python
# Eksempel 9: Karakterbog som dictionary
karakterer = {
    'Anna': 10,
    'Bert': 7,
    'Cecilia': 12,
    'David': 4
}

# Find Annas karakter
print('Annas karakter: ' + str(karakterer['Anna']))

# Find gennemsnit
karakterer_liste = list(karakterer.values())
gennemsnit = sum(karakterer_liste) / len(karakterer_liste)
print('\nGennemsnits karakter: ' + str(gennemsnit))

# Find elever der har fået 10 eller mere
print('\nElever med 10 eller mere:')
for navn, karakter in karakterer.items():
    if karakter >= 10:
        print(navn + ': ' + str(karakter))
````

**Hvad skete her?**
- Vi brugte `.values()` metoden til at få alle værdier som en liste
- Vi kombinerede dictionary med `if` statement
- Vi filtrede og printede kun de elever der fik 10 eller mere

Bemærk: Dictionaries bliver meget mægtige når du kombinerer dem med løkker og betingelser!

## Eksempel 10: Sikker adgang med .get()

Der er en fare når du bruger `dict['noekle']` - hvis nøglen ikke eksisterer, får du en fejl. Her er en sikrere måde:

````python
# Eksempel 10: Sikker adgang med .get()
person = {'navn': 'Anna', 'alder': 12}

# FARLIGT: Dette giver fejl hvis 'email' ikke eksisterer
# print(person['email'])  # Fejl!

# SIKKERT: .get() giver en standard værdi hvis nøglen ikke eksisterer
email = person.get('email', 'ikke opgivet')
print('Email: ' + email)

# Hvis nøglen eksisterer, får vi værdien
navn = person.get('navn', 'Ukendt')
print('Navn: ' + navn)
````

**Hvad skete her?**
- `.get('email', 'ikke opgivet')` betyder: "Prøv at få værdien for 'email', og hvis den ikke eksisterer, giv mig 'ikke opgivet' i stedet"
- Dette undgår fejl hvis du ikke er sikker på at nøglen eksisterer
- Det er en god vane at bruge `.get()` når du arbejder med dictionaries

## Eksempel 11: Tjek hvis nøgle eksisterer

Hvis du vil vide om en nøgle eksisterer i ordbogen, kan du bruge `in`:

````python
# Eksempel 11: Tjek hvis nøgle eksisterer
person = {'navn': 'Anna', 'alder': 12, 'by': 'København'}

# Tjek hvis 'navn' eksisterer
if 'navn' in person:
    print('Navn eksisterer: ' + person['navn'])

# Tjek hvis 'email' eksisterer
if 'email' in person:
    print('Email eksisterer: ' + person['email'])
else:
    print('Email eksisterer ikke')
````

**Hvad skete her?**
- Vi brugte `if 'navn' in person:` til at tjekke hvis nøglen eksisterer
- Hvis den gør, kan vi sikkert få værdien
- Hvis den ikke gør, udskriver vi en besked i stedet

Dette er endnu en måde at undgå fejl når du arbejder med dictionaries.

## Fælles fejl 🐛

### Fejl 1: Glemmer at nøglen skal være en streng

**FORKERT:**
```python
person = {navn: 'Anna', alder: 12}  # FEJL - navn og alder skal være 'navn' og 'alder'
```

**KORREKT:**
```python
person = {'navn': 'Anna', 'alder': 12}  # OK - nøgler er strenge
```

**Forklaring:** Nøgler skal normalt være i citationstegn (være strenge). Hvis du glemmer det, søger Python efter en variabel med det navn i stedet.

---

### Fejl 2: Bruger `[` i stedet for `in` når man tjekker

**FORKERT:**
```python
if person['email']:  # Fejl hvis 'email' ikke eksisterer!
    print('Email findes')
```

**KORREKT:**
```python
if 'email' in person:  # OK - tjekker hvis nøgle eksisterer
    print('Email findes')
```

**Forklaring:** `if person['email']:` giver fejl hvis email ikke eksisterer. Brug `if 'email' in person:` i stedet.

---

### Fejl 3: Glemmer `.items()` når man vil have både nøgler og værdier

**FORKERT:**
```python
for noekle, vaerdi in person:  # Fejl - giver kun nøkler
    print(noekle, vaerdi)
```

**KORREKT:**
```python
for noekle, vaerdi in person.items():  # OK - giver nøgler og værdier
    print(noekle, vaerdi)
```

**Forklaring:** Hvis du løkker gennem dictionaries direkte, får du kun nøklerne. Brug `.items()` hvis du vil have både.

---

### Fejl 4: Blander dictionary med liste-syntaks

**FORKERT:**
```python
ordbog = {'navn': 'Anna', 'alder': 12}
ordbog.append({'by': 'København'})  # FEJL - .append() er for lister!
```

**KORREKT:**
```python
ordbog = {'navn': 'Anna', 'alder': 12}
ordbog['by'] = 'København'  # OK - tilføj til dictionary
```

**Forklaring:** Dictionaries og lister har forskellige metoder. Brug `dict[key] = value` for dictionaries.

---

### Fejl 5: Glemmer kolon mellem nøgle og værdi

**FORKERT:**
```python
person = {'navn' 'Anna', 'alder': 12}  # FEJL - mangler kolon efter 'navn'
```

**KORREKT:**
```python
person = {'navn': 'Anna', 'alder': 12}  # OK
```

**Forklaring:** Mellem nøgle og værdi skal være `:` (kolon). Ikke komma!

---

### Fejl 6: Forveksler dictionary med liste

**FORKERT:**
```python
# Når du skal bruge dictionary
frugter = ['banan', 'æble', 'appelsin']  # Liste - kan ikke få data på navn
pris = frugter['banan']  # FEJL!
```

**KORREKT:**
```python
# Når du har navn-værdi data
frugter = {'banan': 5, 'æble': 7, 'appelsin': 8}  # Dictionary
pris = frugter['banan']  # OK - giver 5
```

**Forklaring:** 
- **Liste** når du har en **samling af samme type** (dyr, tal, ord) og rækkefølge betyder noget
- **Dictionary** når du har **par af information** (navn→nummer, ord→betydning, frugt→pris)
