## Debugging Tips 🐛

### Hvis der er fejl, prøv denne checklist:

**1. Syntaks-fejl**
- Mangler kolon efter `if`, `while`, `for`, `def`?
- Mangler indentation indenfor block?
- Mismatched parenteser/brackets?

**2. Logik-fejl (koden kører men gør det forkert)**
- Bruger du `=` i stedet for `==` i if-statement?
- Er dit while-loop uendeligt? (betingelse bliver aldrig falsk)
- Glemmer du at ændre variabel indenfor løkke?

**3. Input-fejl**
- Input fra bruger er altid en **string**!
- Hvis du skal sammenligne med tal: `int(input())`
- Hvis bruger skrev mellemrum: `.strip()`

**4. Brug print() til at debug**
```python
# Hvis du er forvirret:
print('Her er vi') # Tjek hvis koden når hertil
print('Tal:', tal) # Se hvad variablen indeholder
print('Type:', type(tal)) # Se hvad type den er
```
